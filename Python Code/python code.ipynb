```python
import pandas as pd
df = pd.read_csv('Data Sources\Our World in Data- Grouped GHG.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

# Creating a dataframae for just China
china_df = df[df['country'] == 'China']

# Needed to sort the data by year
china_df = china_df.sort_values(by='year')

# Plotting the data
china_df.plot(x='year', y='total_ghg', kind='line', marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('GHG Emissions')
plt.title('China GHG Emissions Over Time')

# Showing the plot
plt.show()
```

```python
# Creating a dataframe just for the United States
us_df = df[df['country'] == 'United States']

# Needed to sort the data by year for smooth line
us_df = us_df.sort_values(by='year')

# Plotting the data
us_df.plot(x='year', y='total_ghg', kind='line', marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('GHG Emissions')
plt.title('United States GHG Emissions Over Time')

# Showing the plot
plt.show()
```

```python
df_2020 = df[df['year'] == 2020]

# Create pie chart
plt.pie(df_2020['total_ghg'], labels=df_2020['country'])

# Add title
plt.title('GHG Total by Country')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show plot
plt.show()
```

```python
new_df = pd.DataFrame(columns=['country', 'year', 'total_ghg', 'percent_of_total'])

for year in range(2010, 2021):
    df_year = df[df['year'] == year]
    
    total_ghg_sum = df_year['total_ghg'].sum()
    
    df_year['percent_of_total'] = df_year['total_ghg'] / total_ghg_sum
    
    new_df = pd.concat([new_df, df_year], ignore_index=True)

print(new_df)
```

```python
df_other_countries = new_df[new_df['country'] == "All Other Countries"]

print(df_other_countries)
```
